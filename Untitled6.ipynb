```python
#keyboard variable length argument
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a = 10, b = 20, c = 30)
```

```python
def pass_student(**kwargs):
  pass_student_list = []
  for name,marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
    return pass_student_list
```

```python
#non keyword variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2,3,4,5,6)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=sum
  return sum
```

```python
sum_number(1,2,3,4,5,6)
```

```python
cgpa = np.array([6,6.5,7,7.8,8,8.6,9,9.2,9.5,10])
package = np.array([4,4.5,5,5.8,6.2,6.8,7,7.5,8,8.8])
font1 = {'family': 'serif','color': 'blue','size': 15}
font2 = {'family': 'serif','color': 'darkred','size' :20}
plt.plot(cgpa,package,"*:g",ms =15,mec = "Orange", mfc= "Red", linewidth = 2.4)
plt.xlabel("CGPA", fontdict = font1)
plt.ylabel("PACKAGE",fontdict = font1)
plt.title("CGPA VS PACKAGE",fontdict = font2)
plt.show()
```

```python
import numpy as np
from numpy.linalg import norm

A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])

#compute cosinne similarity
cosine = np.dot(A,B) / (norm(A)* norm(B))
print("Cosine Similarity:",cosine)
```
